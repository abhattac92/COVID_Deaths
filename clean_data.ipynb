```python
# calculate days between deadline and creation
data_clean[['created_at','deadline','launched_at']] = data_clean[['created_at','deadline','launched_at']].apply(pd.to_datetime) #if conversion required
data_clean['launch_to_deadline'] = (data_clean['deadline'] - data_clean['launched_at']).dt.days
data_clean['creation_to_launch'] = (data_clean['launched_at']- data_clean['created_at']).dt.days
```

```python
# day of the week for launch, deadline
data_clean['launch_day'] = data_clean['launched_at'].dt.day_name()
data_clean['deadline_day'] = data_clean['deadline'].dt.day_name()
# month
data_clean['launch_month'] = data_clean['launched_at'].dt.month_name()
data_clean['deadline_month'] = data_clean['deadline'].dt.month_name()
```

```python
data_clean
```

```python
# Launch time
data_clean['launch_hour'] = data_clean['launched_at'].dt.hour 

def two_hour_bucket(row):
    if row['launch_hour'] in (0,1):
        return '12am-2am'
    if row['launch_hour'] in (2,3):
        return '2am-4am'
    if row['launch_hour'] in (4,5):
        return '4am-6am'
    if row['launch_hour'] in (6,7):
        return '6am-8am'
    if row['launch_hour'] in (8,9):
        return '8am-10am'
    if row['launch_hour'] in (10,11):
        return '10am-12pm'
    if row['launch_hour'] in (12,13):
        return '12pm-2pm'
    if row['launch_hour'] in (14,15):
        return '2pm-4pm'
    if row['launch_hour'] in (16,17):
        return '4pm-6pm'
    if row['launch_hour'] in (18,19):
        return '6pm-8pm'
    if row['launch_hour'] in (20,21):
        return '8pm-10pm'
    if row['launch_hour'] in (22,23):
        return '10pm-12am'
    
data_clean['launch_time'] = data_clean.apply(two_hour_bucket, axis=1) 
data_clean.drop('launch_hour', axis=1, inplace=True)
```

```python
# deadline time
data_clean['deadline_hour'] = data_clean['deadline'].dt.hour 

def two_hour_bucket(row):
    if row['deadline_hour'] in (0,1):
        return '12am-2am'
    if row['deadline_hour'] in (2,3):
        return '2am-4am'
    if row['deadline_hour'] in (4,5):
        return '4am-6am'
    if row['deadline_hour'] in (6,7):
        return '6am-8am'
    if row['deadline_hour'] in (8,9):
        return '8am-10am'
    if row['deadline_hour'] in (10,11):
        return '10am-12pm'
    if row['deadline_hour'] in (12,13):
        return '12pm-2pm'
    if row['deadline_hour'] in (14,15):
        return '2pm-4pm'
    if row['deadline_hour'] in (16,17):
        return '4pm-6pm'
    if row['deadline_hour'] in (18,19):
        return '6pm-8pm'
    if row['deadline_hour'] in (20,21):
        return '8pm-10pm'
    if row['deadline_hour'] in (22,23):
        return '10pm-12am'
    
data_clean['deadline_time'] = data_clean.apply(two_hour_bucket, axis=1) 
data_clean.drop('deadline_hour', axis=1, inplace=True)
```
