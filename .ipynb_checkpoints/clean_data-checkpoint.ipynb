```python
# dropping columns with small no-null values
clean.drop(['friends', 'is_backing', 'is_starred', 'permissions','last_update_published_at','unread_messages_count','unseen_activity_count'], axis=1, inplace=True)
```

```python
# Dropping useless columns
clean.drop(['converted_pledged_amount', 'country', 'creator', 'currency', 'currency_symbol', 'currency_trailing_code','disable_communication', 'fx_rate', 'is_starrable', 'name', 'photo', 'profile', 'slug', 'source_url', 'spotlight', 'state_changed_at', 'static_usd_rate','urls', 'usd_type', 'current_currency'], axis=1, inplace=True)
```

```python
clean.head()
```

need to convert unix times to standard

```python
# Converting dates from unix to datetime
unix_convert = ['created_at', 'deadline', 'launched_at']
for x in unix_convert:
    clean[x] = pd.to_datetime(clean[x], origin='unix', unit='s')
```

```python
clean.head()
```

```python
# Extract sub-category section from category
f = lambda x: x['category'].split('/')[1].split('","position')[0]
clean['sub_category'] = clean.apply(f, axis=1)

#extract and replace
f = lambda x: x['category'].split('"slug":"')[1].split('/')[0]
clean['category'] = clean.apply(f, axis=1)
f = lambda x: x['category'].split('","position"')[0] 
clean['category'] = clean.apply(f, axis=1)
```

```python
clean.head()
```

```python
# Counting the number of unique countries
clean.country_displayable_name.unique()
```

```python
#location column has way too many unique values, will drop

clean.drop('location', axis=1, inplace=True)
```

```python
clean.head()
```

blurb
We will not be focusing on NLP for this project so we will convert this info into something relevant for us. We will look at the length of the blurbs and see if this is has any impact

```python
# Count length of each blurb
clean['blurb_length'] = clean['blurb'].str.split().str.len()

# Drop blurb variable
clean.drop('blurb', axis=1, inplace=True)
```
